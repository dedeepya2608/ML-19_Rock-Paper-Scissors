```python
from keras.models import load_model
model= load_model("rock_paper_scissors-model.h5")
```

```python
categories = ['rock', 'paper', 'scissors']
```

```python
import cv2
from random import choice
import numpy as np
import os
def calculate_winner(move1, move2):
    if move1 == move2:
        return "Tie"

    if move1 == "rock":
        if move2 == "scissors":
            return "User"
        if move2 == "paper":
            return "Computer"

    if move1 == "paper":
        if move2 == "rock":
            return "User"
        if move2 == "scissors":
            return "Computer"

    if move1 == "scissors":
        if move2 == "paper":
            return "User"
        if move2 == "rock":
            return "Computer"
        
cap = cv2.VideoCapture(0)


while True:
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)
    if not ret:
        continue

    # rectangle for user to play
    cv2.rectangle(frame, (50, 50), (250, 250), (255, 255, 255), 2)
    # rectangle for computer to play
    cv2.rectangle(frame, (400, 50), (600, 250), (255, 255, 255), 2)

    # extract the region of image within the user rectangle
    roi = frame[50:250, 50:250]
    img = cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (224, 224))

    # predict the move made
    pred = model.predict(np.array([img]))
    pred_user_move = np.argmax(pred[0]) # code 0,1,2 predicted by computer
    user_move_name= categories[pred_user_move]
    computer_move_name = choice(['rock', 'paper', 'scissors'])
    winner = calculate_winner(user_move_name, computer_move_name)
    
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame, "Your Move: " + user_move_name,
                (25, 25), font, 1, (255, 255, 255), 2, cv2.LINE_AA)
    cv2.putText(frame, "Computer's Move: " + computer_move_name,
                (375, 25), font, 1, (255, 255, 255), 2, cv2.LINE_AA)
    cv2.putText(frame, "Winner: " + winner,
                (200, 300), font, 1.2, (0, 0, 255), 4, cv2.LINE_AA)
    
    
   # for out_img in os.listdir('C:\\Users\\Dedeepya\\Desktop\\output'):
        #out_img_path = os.path.join('C:\\Users\\Dedeepya\\Desktop\\output',out_img)
        
    output_array= cv2.imread(os.path.join('C:\\Users\\Dedeepya\\Desktop\\output',computer_move_name))
    #output_array = cv2.resize(output_array,(400,400))
    frame[100:500, 800:1200] = output_array
    
    
    
    
    
    
    
    
    
    

    cv2.imshow("Rock Paper Scissors", frame)

    k = cv2.waitKey(10)
    if k == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
    
```
